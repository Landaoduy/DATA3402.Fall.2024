1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter.

```python
class Counter:
    
    def __init__(self, initial_value, maximum_value):
        self.initial_value = initial_value
        self.maximum_value = maximum_value
        self.current_value = initial_value
    
    def increment(self):
        
        if self.current_value < self.maximum_value:
            self.current_value += 1
            print(self.current_value)
            
        else:
            print("Error: counter has reached its maximum value")
            
            
    def reset(self):
        self.current_value = self.initial_value
        print("Counter reset to:", self.current_value)
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    
    def __init__(self, name): 
        self.name = name
        
    def get_name(self):
        print(f"The object is a: {self.name}")
    
    @abstractmethod
    def is_the_object_overlap(self, other):
        pass


class Rectangle(Shape):
    
    def __init__(self, name, length, width, x, y): 
        super().__init__(name)
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
    
    # Accessors
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def get_x_coordinate(self):
        return self.__x
    
    def get_y_coordinate(self):
        return self.__y

    def get_top_right(self):
        return (self.__x + self.__length, self.__y + self.__width)
    
    def get_bottom_left(self):
        return (self.__x, self.__y)
     
    
    def is_the_object_overlap(self, other):
        
        l1, r1 = self.get_bottom_left(), self.get_top_right()
        l2, r2 = other.get_bottom_left(), other.get_top_right()
        
        if l1[0] > r2[0] or l2[0] > r1[0]:
            return False

        if l1[1] > r2[1] or l2[1] > r1[1]:
            return False

        return True

class Circle(Shape):
    
    
    def __init__(self, name, radius, x, y):
        super().__init__(name)
        self.__radius = radius
        self.__x = x
        self.__y = y
        
    # Accessors
    
    def get_radius(self):
        return self.__radius
    
    def get_x_coordinate(self):
        return self.__x
    
    def get_y_coordinate(self):
        return self.__y
  
    def center(self):
        return (self.__x, self.__y)
  
    def is_the_object_overlap(self, other):
        
        c1, r1 = self.center(), self.get_radius()
        c2, r2 = other.center(), other.get_radius()

        distance = ((c1[0] - c2[0])**2 + (c1[1] - c2[1])**2)**0.5
        
        if distance > r1 + r2:
            return False # Not overlap

        return True # Overlap

class Triangle(Shape):
    
    def __init__(self, name, vertex1, vertex2, vertex3):
        super().__init__(name)
        self.__vertex1 = vertex1 # tuple (x1, y1)
        self.__vertex2 = vertex2 # tuple (x2, y2)
        self.__vertex3 = vertex3 # tuple (x3, y3)
    
    # Accessors
    def get_vertices(self):
        
        return [self.__vertex1, self.__vertex2, self.__vertex3]
    
    def bounding_box(self):
        
        xs = [v[0] for v in self.get_vertices()]
        ys = [v[1] for v in self.get_vertices()]
        
        min_x, max_x = min(xs), max(xs)
        min_y, max_y = min(ys), max(ys)
        
        return (min_x, min_y), (max_x, max_y) # bottom-left and top-right corners
    
    def is_the_object_overlap(self, other):
        
        if isinstance(other, Triangle):
            
            l1, r1 = self.bounding_box()
            l2, r2 = other.bounding_box()
            
            if l1[0] > r2[0] or l2[0] > r1[0]:
                return False
            
            if l1[1] > r2[1] or l2[1] > r1[1]:
                return False
            
            return True
        
        return False
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]
    
    def v_line(self, x, y, w, **kwargs):
        char = kwargs.get('char', '*')
        for i in range(w):
            self.set_pixel(x + i, y, char)

    def h_line(self, x, y, h, **kwargs):
        char = kwargs.get('char', '*')
        for i in range(h):
            self.set_pixel(x, y + i, char)

    def line(self, x1, y1, x2, y2, **kwargs):
        char = kwargs.get('char', '*')
        dx = abs(x2 - x1)
        dy = abs(y2 - y1)
        sx = 1 if x1 < x2 else -1
        sy = 1 if y1 < y2 else -1
        err = dx - dy

        while True:
            self.set_pixel(x1, y1, char)
            if x1 == x2 and y1 == y2:
                break
            e2 = 2 * err
            if e2 > -dy:
                err -= dy
                x1 += sx
            if e2 < dx:
                err += dx
                y1 += sy
    
    def display(self):
        for row in self.data:
            print(''.join(row))

class Shape:
    def __init__(self, name="", **kwargs):
        self.name = name
        self.kwargs = kwargs
    
    def paint(self, canvas): 
        pass

class Rectangle(Shape):
    def __init__(self, x, y, w, h, **kwargs):
        super().__init__(**kwargs)
        self.x = x
        self.y = y
        self.w = w
        self.h = h
    def paint(self, canvas):
        canvas.v_line(self.x, self.y, self.w, **self.kwargs)
        canvas.v_line(self.x, self.y + self.h, self.w, **self.kwargs)
        canvas.h_line(self.x, self.y, self.h, **self.kwargs)
        canvas.h_line(self.x + self.w, self.y, self.h, **self.kwargs)

class Circle(Shape):
    def __init__(self, x, y, radius, **kwargs):
        super().__init__(**kwargs)
        self.radius = radius
        self.x = x
        self.y = y
    
    def paint(self, canvas):
        for angle in range(0, 360, 11): 
            rad = math.radians(angle)
            col = int(self.x + self.radius * math.cos(rad))
            row = int(self.y + self.radius * math.sin(rad))
            if 0 <= row < canvas.height and 0 <= col < canvas.width:
                canvas.set_pixel(row, col, **self.kwargs)

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3, **kwargs):
        super().__init__(**kwargs)
        self.x1, self.y1 = x1, y1
        self.x2, self.y2 = x2, y2
        self.x3, self.y3 = x3, y3
    
    def paint(self, canvas):
        canvas.line(self.x1, self.y1, self.x2, self.y2, **self.kwargs)
        canvas.line(self.x2, self.y2, self.x3, self.y3, **self.kwargs)
        canvas.line(self.x3, self.y3, self.x1, self.y1, **self.kwargs)
    
class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes
    
    def paint(self, canvas):
        for s in self.shapes:
            s.paint(canvas)
```

```python
class RasterDrawing:
    def __init__(self):
        self.shapes=dict()
        self.shape_names=list()
        
    def add_shape(self,shape):
        if shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name]=shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self,canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base="shape"
        name = name_base+"_0"
        
        i=1
        while name in self.shapes:
            name = name_base+"_"+str(i)
            
        return name
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]
    
    def v_line(self, x, y, w, **kwargs):
        char = kwargs.get('char', '*')
        for i in range(w):
            self.set_pixel(x + i, y, char)

    def h_line(self, x, y, h, **kwargs):
        char = kwargs.get('char', '*')
        for i in range(h):
            self.set_pixel(x, y + i, char)

    def line(self, x1, y1, x2, y2, **kwargs):
        char = kwargs.get('char', '*')
        dx = abs(x2 - x1)
        dy = abs(y2 - y1)
        sx = 1 if x1 < x2 else -1
        sy = 1 if y1 < y2 else -1
        err = dx - dy

        while True:
            self.set_pixel(x1, y1, char)
            if x1 == x2 and y1 == y2:
                break
            e2 = 2 * err
            if e2 > -dy:
                err -= dy
                x1 += sx
            if e2 < dx:
                err += dx
                y1 += sy
    
    def display(self):
        for row in self.data:
            print(''.join(row))

class Shape:
    def __init__(self, name="", **kwargs):
        self.name = name
        self.kwargs = kwargs
    
    def __repr__(self):
        raise NotImplementedError
    
    def paint(self, canvas): 
        pass

class Rectangle(Shape):
    def __init__(self, x, y, w, h, **kwargs):
        super().__init__(**kwargs)
        self.x = x
        self.y = y
        self.w = w
        self.h = h

    def paint(self, canvas):
        canvas.v_line(self.x, self.y, self.w, **self.kwargs)
        canvas.v_line(self.x, self.y + self.h, self.w, **self.kwargs)
        canvas.h_line(self.x, self.y, self.h, **self.kwargs)
        canvas.h_line(self.x + self.w, self.y, self.h, **self.kwargs)
    
    def __repr__(self):
        return f"Rectangle({self.x}, {self.y}, {self.w}, {self.h}, name={repr(self.name)}, **{repr(self.kwargs)})"

class Circle(Shape):
    def __init__(self, x, y, radius, **kwargs):
        super().__init__(**kwargs)
        self.radius = radius
        self.x = x
        self.y = y
    
    def paint(self, canvas):
        for angle in range(0, 360, 11): 
            rad = math.radians(angle)
            col = int(self.x + self.radius * math.cos(rad))
            row = int(self.y + self.radius * math.sin(rad))
            if 0 <= row < canvas.height and 0 <= col < canvas.width:
                canvas.set_pixel(row, col, **self.kwargs)
    
    def __repr__(self):
        return f"Circle({self.x}, {self.y}, {self.radius}, name={repr(self.name)}, **{repr(self.kwargs)})"

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3, **kwargs):
        super().__init__(**kwargs)
        self.x1, self.y1 = x1, y1
        self.x2, self.y2 = x2, y2
        self.x3, self.y3 = x3, y3
    
    def paint(self, canvas):
        canvas.line(self.x1, self.y1, self.x2, self.y2, **self.kwargs)
        canvas.line(self.x2, self.y2, self.x3, self.y3, **self.kwargs)
        canvas.line(self.x3, self.y3, self.x1, self.y1, **self.kwargs)
    
    def __repr__(self):
        return f"Triangle({self.x1}, {self.y1}, {self.x2}, {self.y2}, {self.x3}, {self.y3}, name={repr(self.name)}, **{repr(self.kwargs)})"
    
class RasterDrawing:
    
    def __init__(self):
        self.shapes=dict()
        self.shape_names=list()
        
    def add_shape(self,shape):
        
        if shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name]=shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self,canvas):
        
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        
        name_base="shape"
        name = name_base+"_0"
    
        i=1
        while name in self.shapes:
            name = name_base+"_"+str(i)
            
        return name
  
    def save(self, filename):
        with open(filename, "w") as file:
            for shape_name in self.shape_names:
                file.write(repr(self.shapes[shape_name]) + "\n")

    def load(self, filename):
        self.shapes.clear()
        self.shape_names.clear()
        
        with open(filename, "r") as file:
            
            for line in file:
                shape = eval(line.strip()) 
                self.add_shape(shape)
```
